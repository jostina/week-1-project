```python
# 1. Find all records where the weather was exactly clear.
clear_weather_records = df[df['Weather'] == 'Clear']
print("Records where the weather was exactly clear:")
print(clear_weather_records)
```

```python
# 2. Find the number of times the wind speed was exactly 4 km/hr.
wind_speed_4_count = df[df['Wind Speed_km/h'] == 4].shape[0]
print(f"Number of times the wind speed was exactly 4 km/hr: {wind_speed_4_count}")
```

```python
# 3. Check if there are any NULL values present in the dataset.
null_values = df.isnull().sum()
print("NULL values present in the dataset:")
print(null_values)
```

```python
# 4. Rename the column "Weather" to "Weather_Condition."
df.rename(columns={'Weather': 'Weather_Condition'}, inplace=True)
print("Columns after renaming 'Weather' to 'Weather_Condition':")
print(df.columns)
```

```python
# 5. What is the mean visibility of the dataset?
mean_visibility = df['Visibility_km'].mean()
print(f"Mean visibility of the dataset: {mean_visibility}")
```

```python
# 6. Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.
high_wind_low_visibility_records = df[(df['Wind Speed_km/h'] > 24) & (df['Visibility_km'] == 25)].shape[0]
print(f"Number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km: {high_wind_low_visibility_records}")
```

```python
# 7. What is the mean value of each column for each weather condition?
# Exclude the 'Date/Time' column as it's not numeric
mean_values_per_weather_condition = df.drop('Date/Time', axis=1).groupby('Weather_Condition').mean()
print("Mean value of each column for each weather condition:")
print(mean_values_per_weather_condition)
```

```python
# 8. Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.
specific_conditions = df[(df['Weather_Condition'] == 'Clear') & (df['Rel Hum_%'] > 50) | (df['Visibility_km'] > 40)]
print("Instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40:")
print(specific_conditions)
```

```python
# 9. Find the number of weather conditions that include snow.
snow_conditions_count = df[df['Weather_Condition'].str.contains('Snow', case=False, na=False)].shape[0]
print(f"Number of weather conditions that include snow: {snow_conditions_count}")
```
